```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
import math

```

```python
file_obj = open('answer1.txt', 'w')
def f(x):
    return math.sin(x/5.)*math.exp(x/10.)+5*math.exp(-x/2.)

x = np.arange (0., 31.)
y = np.array(x)
for i in xrange(0, x.size):
    y[i] = f(x[i])

res = str(opt.minimize(f, 2, method='BFGS'))
file_obj.writelines(res)
plt.plot (x , y)
```
